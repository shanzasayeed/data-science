Create classification dataset

```python
from sklearn.datasets import make_classification 
```

```python
#x=independent feature
#y=dependent feature that is target
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)       
```

```python
x
```

```python
y
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
```

```python
df1
```

```python
df2=pd.DataFrame(y,columns=['target'])
```

```python
df2
```

```python
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#transform the dataset
over_sample=SMOTE()
x,y=over_sample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
x.shape
```

```python
y.shape
```

```python
y.value_counts()
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])

oversampled_df=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversampled_df['f1'],oversampled_df['f2'],c=oversampled_df['target'])
```
