Scatter Plot

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
#c=colours
colours=np.random.rand(50)
plt.scatter(x,y,c=colours)
```

```python
#instead of different colours we can give any particular colour as well
plt.scatter(x,y,c='b')
```

```python
plt.scatter(x,y,c='r')
```

```python
# you can see html colour code on google
plt.scatter(x,y,c='#3CB371')
```

```python
plt.scatter(x,y,marker='x')
```

```python
plt.scatter(x,y,marker='v')
```

```python
plt.scatter(x,y,marker='o')
```

```python
#s=size
sizes=np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes)
```

```python
sizes=100*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes)
```

```python
#alpha is intensity of colour
sizes=100*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.5)
```

```python
plt.scatter(x,y,c=colours,s=sizes,alpha=.3)
```

```python
plt.scatter(x,y,c=colours,s=sizes,alpha=.3)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
```

3d projection

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot(x,y,z)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.bar(x,y,z)
plt.show()
```
