```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

graph - representation of data on axis

```python
x=np.linspace(0,10,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
#labelling x and y data ,giving title and showing graph
plt.plot(x,y)
plt.xlabel('this is my x data')   
plt.ylabel('this is my y data')
plt.title('This is my title')
plt.show()
```

```python
plt.plot(x,y,'--r')  # dotted red line
plt.xlabel('this is my x data')   
plt.ylabel('this is my y data')
plt.title('This is my title')
plt.show()
```

```python
plt.plot(x,y,'--b')  # dotted blue line
plt.xlabel('this is my x data')   
plt.ylabel('this is my y data')
plt.title('This is my title')
plt.show()
```

```python
#managing size of figure
plt.figure(figsize=(9,5))  #size of x axis and y axis
plt.plot(x,y,'--r')  
plt.xlabel('this is my x data')   
plt.ylabel('this is my y data')
plt.title('This is my title')
plt.show()
```

```python
plt.figure(figsize=(9,2))  #size of x axis and y axis
plt.plot(x,y,'--r')  
plt.xlabel('this is my x data')   
plt.ylabel('this is my y data')
plt.title('This is my title')
plt.grid()  #get grid in background
plt.show()
```

SCATTER GRAPH

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.scatter(x,y, )
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
colours=np.random.rand(50)
plt.scatter(x,y, c = colours)    # to give different colours to each point
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
colours=np.random.rand(50)
plt.scatter(x,y, c = 'b')    # to give specified colours to each point (b is blue)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
colours=np.random.rand(50)
plt.scatter(x,y, c = '#FFFF00')   # can also give html colour code of any colour(from google)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
sizes=100 * np.random.rand(50)
colours=np.random.rand(50)
plt.scatter(x,y, c = colours,s=sizes)    # to give different sizes to each point
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
sizes=100 * np.random.rand(50)
colours=np.random.rand(50)
plt.scatter(x,y, c = colours,s=sizes,alpha=.5)    # alpha is ntensity of coour
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph')
plt.show()
```

```python
plt.scatter(x,y,marker='x') # to get a cross instead of circle
```

```python
plt.scatter(x,y,marker='v')
```

BAR GRAPH

```python
x=['a','b','c','d','e'] #string value(categorical value)
```

```python
y=np.random.rand(5)   #numeric value
```

```python
plt.bar(x,y) #when we have 1 categorical value on  axis and numerical value on other axis
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is bar graph')
plt.show() 
```

```python
#bar data in horizontal mode(horizontal bar)
plt.barh(x,y)
```

3-D representation

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot(x,y,z)
plt.show()
```
